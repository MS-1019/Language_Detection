## converting first text to array and then in vectors for machine understandable code...

```python
## converting to on=bject or text to list 
X=np.array(Lang['Text'])
Y=np.array(Lang['Language'])
print(X, '\n',Y)
```

```python
Lang.drop_duplicates(keep='first')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
cv=CountVectorizer()
x=cv.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,Y,test_size=0.27,random_state=42)
X_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
M=MultinomialNB()
M.fit(X_train,y_train)
```

```python
M.score(X_test,y_test)
```

```python
input1=input("Enter the text to detect language")
model=cv.transform([input1]).toarray()
predictt=M.predict(model)
print(predictt)
```
